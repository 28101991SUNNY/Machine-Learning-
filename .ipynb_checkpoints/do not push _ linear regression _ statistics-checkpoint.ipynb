    dE/db = d/db [(y-mx-b)^2]
           = d/db [ (y-mx-b)^2 ] = 0
           = 2(y-mx-b) *  d/db[(y-mx-b)]
           = 2(y-mx-b)(-1)
           = -2(y-mx-b) = 0

       partial derivation wrt to m
       
       
       
       
       dE/dm = d/dm [(y-mx-b)^2]
       = d/dm [ (y-mx-b)^2 ] = 0    
       
       = d/dm [ (y-mx-ybar + m xbar)  **2 ] = 0
       
       [ 2(y-ybar-m x +m xbar) * d/dm ((y-mx-ybar +m xbar)) ] = 0
       
       2(y-ybar-mx+ mxbar) * (-x+xbar) = 0
       
       -2(y-mx -ybar+mxbar)(x-xbar)= 0
       
       (y-mx -ybar+mxbar)(x-xbar)= 0
       
       ([y-ybar  - m (x-xbar)]) (x-xbar) = 0
       
       (y-ybar)(x-xbar) -m(x-xbar)^2 = 0
       (y-ybar)(x-xbar) = m(x-xbar)^2 
       
       sum((y-ybar)(x-xbar)) / sum((x-xbar)^2 ) = m
       
       

```python
class simpleLinearRegressionOLS:
    
    def __init__(self):
        self.m = None
        self.b = None
        
    def fit(self,X_train,y_train):
        
        
        numerator = np.sum((X_train - X_train.mean()) * (y_train - y_train.mean()) )
        denominator  = np.sum((X_train-X_train.mean())**2)
            
        self.m = (numerator/denominator)    
        
        self.b = (y_train.mean()) - (self.m * X_train.mean())
        
        print("Weight: ",self.m)
        print("Intercept: ",self.b)
        
    def predict(self,X_test):
        return (X_test * self.m) + (self.b)
    
    def MAE(self):
        '''
        Mean absolute Error
        '''
        return (np.sum(np.abs(y_train-self.predict(X_train))))/(len(X_train))

    def MSE(self):
        '''
        Mean Squared Error
        '''
        return ((np.sum((y_train-self.predict(X_train))**2)))/(len(X_train))
    
    def RMSE(self):
        return np.sqrt(self.MSE())
    
    def r2_score(self):
        '''
        coefficient of determination:
        goodness of fit:
        
        r2score = 1- (SSR/SST)
        SSR : sum of squared error in regression line
        SST : sum of squared error in mean line
        '''
        y_pred = self.predict(X_test)
        return 1- (np.sum((y_test - y_pred)**2)  /np.sum((y_test - y_test.mean())**2) )
    
    
    def adjusted_r2_score(self):
        d = np.ndim(X_test)
        n = len(X_test)
        
        return 1- ( ((1-self.r2_score())*(n-1)) /(n-1-d) )
```

```python
class MultiLinearRegressor:
    
    def __init__(self):
        self.coef_ = None
        self.intercept_ = None
    
    def fit(self,X_train, y_train):
        
        X_train = np.insert(X_train,0,1,axis = 1)
        
        # calculate coefficients : b = (xTx)-1 . xT.y
        all_betas = np.dot(np.linalg.inv(np.dot(X_train.T,X_train)),np.dot(X_train.T,y_train))
        
        self.coef_ = all_betas[1:]
        self.intercept_ = all_betas[0]
        
        
        print(all_betas[0])
        return (all_betas[1:])
        
    def predict(self,X_test):
        
         return np.dot(X_test,self.coef_)+ self.intercept_
```

```python
lr = MultiLinearRegressor()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
y_pred = lr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
